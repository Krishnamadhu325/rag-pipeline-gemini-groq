```python
from langchain_google_genai import GoogleGenerativeAIEmbeddings
from langchain_community.vectorstores import Chroma
```

```python
embeddings=GoogleGenerativeAIEmbeddings(
    model='models/text-embedding-004',api_key=GEMINI_API_KEY)
```

```python
texts = [
    "Airbus is a global leader in aircraft manufacturing.",
    "The A320 family is one of the most successful jetliners.",
    "The A350 is known for its fuel efficiency and long range.",
    "Airbus also develops helicopters through Airbus Helicopters.",
    "LangChain makes it easier to build applications powered by LLMs.",
    "Chroma is an open-source vector database for storing embeddings.",
    "Google Generative AI provides high-quality embedding models.",
    "Machine learning models require feature vectors for similarity search.",
    "The aerospace industry relies heavily on digital twin technology.",
    "Predictive maintenance helps airlines reduce downtime and cost.",
]
```

```python
db=Chroma.from_texts(texts,embedding=embeddings)
query="Tell me something about the Airbus aircraft."

results = db.similarity_search(query,k=3)

for r in results:
  print("- ",r.page_content)
```
