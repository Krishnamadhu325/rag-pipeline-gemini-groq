```python
from google.colab import userdata
GROQ_API_KEY = userdata.get('GROQ_API_KEY')
```

```python
from langchain_community.document_loaders import PyPDFLoader
```

```python
pdf_path = "/content/drive/MyDrive/GEN_AI/access-scholarship-rules.pdf"
loader=PyPDFLoader(pdf_path)
docs=loader.load()
```

```python
print(f"Loaded {len(docs)} pages")
```

```python
print("Metadata of 1st page:\n",docs[0].metadata)
```

```python
print("Text content of first page:\n",docs[0].page_content[:2500])
```

```python
from langchain_groq import ChatGroq
llm = ChatGroq(model="openai/gpt-oss-20b",api_key=GROQ_API_KEY)
sample_data=docs[0].page_content

res = llm.invoke(f"Summarize the data {sample_data}")
print(res.content)
```
